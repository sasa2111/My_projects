1 задача. С помощью мультипликативного метода формирования случайной последовательности, построим последовательность случайных чисел (100). Первый член последовательности X0 задаётся произвольным образом.

```python
#в задаче не указан интервал, куда падают случайные числа.
#Предположим, что на интервале [0, 1] надо построить 100 случайных чисел.
#(не сложно это сделать на любом в общем-то отрезке, масштабированием)

х0 = 145729 #произвольное число
t = 5 #пусть такое
lamb = 8 * t + 3
p = 10
q = 6
M = p ** q
random_list = []
for i in range(1,100):
  x1 = lamb * х0 % M
  random_list.append(x1 / M)
  х0 = x1

print(random_list)
```

```python
import matplotlib.pyplot as plt
#посмотрим на графиках, похоже ли распределение на равномерное
```

```python
plt.scatter(range(1,100), random_list); #чисто визуально что-то похожее
```

```python
plt.hist(random_list, bins = 100); #здесь тоже видим, что гистограмма не отдает особого приоритета каким-то отдельным интервалам
```

```python


х0 = 145729 #произвольное число
t = 5 #пусть такое
lamb = 8 * t + 3
p = 10
q = 6
M = p ** q
random_list1 = []
for i in range(0,10):
  x1 = lamb * х0 % M
  random_list1.append(-log(1 - x1/M) / 4)
  х0 = x1

print(random_list1)
```

```python
х0 = 145729 #произвольное число
t = 5 #пусть такое
lamb = 8 * t + 3
p = 10
q = 6
M = p ** q
random_list1 = []
for i in range(0,1000):
  x1 = lamb * х0 % M
  random_list1.append(-log(1 - x1/M) / 4)
  х0 = x1

print(random_list1)
```

```python
plt.scatter(range(0,1000), random_list1); #вот это уже больше похоже, видим концентрацию околонулевых значений
```
